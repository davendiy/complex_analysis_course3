```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(5) + 1j*np.arange(5)
fig, ax = plt.subplots()
ax.scatter(a.real, a.imag)

plt.plot(a.real, a.imag)
plt.show()
```

```python
def region1(z):
    
    if hasattr(z, "__iter__"):
        return np.array([_z for _z in z if _region1(_z)])
    else:
        return np.array([z]) if _region1(z) else np.array([]) 

def _region1(z):
    return 2 < np.abs(z) and np.abs(z - (1j*-4)) < 4


x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)

xxyy = np.array([_x + 1j*_y for _x in x for _y in y])
xxyy = region1(xxyy)

fig, ax = plt.subplots()
ax.scatter(xxyy.real, xxyy.imag, s=1)
```
